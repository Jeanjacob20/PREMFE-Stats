```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Problem 1(a): Joint PDF f(x,y) = 3(x^2 + y)/11
x = np.linspace(0, 2, 50)
y = np.linspace(0, 1, 50)
X, Y = np.meshgrid(x, y)
Z = 3 * (X**2 + Y) / 11

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
ax.set_title('Problem 1(a): Joint PDF')
plt.tight_layout()
plt.savefig('problem1_a.png', dpi=150)
plt.show()
```

```python
# Problem 1(b): Marginal PDF of X
x_vals = np.linspace(0, 2, 100)
fx = 3 * (2 * x_vals**2 + 1) / 22

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x_vals, fx, 'b-', linewidth=2)
ax.set_xlabel('x')
ax.set_ylabel('f_X(x)')
ax.set_title('Problem 1(b): Marginal PDF of X')
ax.grid(True)
plt.tight_layout()
plt.savefig('problem1_b.png', dpi=150)
plt.show()
```

```python
# Problem 1(c): Marginal PDF of Y
y_vals = np.linspace(0, 1, 100)
fy = (8 + 6 * y_vals) / 11

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(y_vals, fy, 'r-', linewidth=2)
ax.set_xlabel('y')
ax.set_ylabel('f_Y(y)')
ax.set_title('Problem 1(c): Marginal PDF of Y')
ax.grid(True)
plt.tight_layout()
plt.savefig('problem1_c.png', dpi=150)
plt.show()
```

```python
# Problem 1(d): Conditional PDF f(y|x) for X = 0, 1, 2
y_vals = np.linspace(0, 1, 100)

fig, ax = plt.subplots(figsize=(6, 4))
for x_val in [0, 1, 2]:
    f_cond = 2 * (x_val**2 + y_vals) / (2 * x_val**2 + 1)
    ax.plot(y_vals, f_cond, linewidth=2, label=f'X = {x_val}')

ax.set_xlabel('y')
ax.set_ylabel('f(y|x)')
ax.set_title('Problem 1(d): Conditional PDF')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.savefig('problem1_d.png', dpi=150)
plt.show()
```

```python
# Problem 1(h): CEF and BLP
x_vals = np.linspace(0, 2, 100)

CEF = (3 * x_vals**2 + 2) / (3 * (2 * x_vals**2 + 1))
BLP = (96 - 10 * x_vals) / 151

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x_vals, CEF, 'b-', linewidth=2, label='CEF: E[Y|X]')
ax.plot(x_vals, BLP, 'r--', linewidth=2, label='BLP: E*(Y|X)')
ax.set_xlabel('X')
ax.set_ylabel('E[Y|X]')
ax.set_title('Problem 1(h): CEF vs BLP')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.savefig('problem1_h.png', dpi=150)
plt.show()
```

```python
# Problem 4: Verify projection matrices
X = np.array([[1, 4], [1, -2], [1, 3], [1, -5]])
XtX = X.T @ X
print("X'X =\n", XtX)

XtX_inv = np.linalg.inv(XtX)
P = X @ XtX_inv @ X.T
M = np.eye(4) - P

print("\nP =\n", P)
print("\nM =\n", M)
print("\nMP =\n", M @ P)
print("\nVerify MP = 0:", np.allclose(M @ P, 0))

# Part (b): X_tilde = XQ
Q = np.array([[1, 3], [2, 8]])
X_tilde = X @ Q
print("\nX_tilde =\n", X_tilde)

P_tilde = X_tilde @ np.linalg.inv(X_tilde.T @ X_tilde) @ X_tilde.T
print("\nP_tilde =\n", P_tilde)
print("\nP_tilde equals P:", np.allclose(P_tilde, P))
```
